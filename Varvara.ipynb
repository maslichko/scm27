### Метод разделимых наименьших квадратов. Квантовые дефекты

Используя метод разделимых наименьших квадратов, построим модель вида
$$
y = \beta_1 - \beta_2 x - \frac{\arctan(\frac{\beta_3}{x-\beta_4})}{\pi}
$$

Предиктор $x$ – возбужденное состояние энергии, 
переменная отклика $y$ – число квантовых дефектов.


```python
import matplotlib.pyplot as plt
import scipy.constants as sc
from scipy.optimize import curve_fit
import numpy as np

def func(x, b1, b2, b3, b4):
    return b1-b2*x-(np.arctan(b3/(x-b4))/sc.pi)

x_data = np.array([-4868.68, -4868.09, -4867.41, -3375.19, -3373.14, -3372.03, -2473.74, -2472.35, -2469.45, -1894.65, -1893.40, -1497.24, -1495.85, -1493.41, -1208.68, -1206.18, -1206.04, -997.92, -996.61, -996.31, -834.94, -834.66, -710.03, -530.16, -464.17])
y_data = np.array([0.252429, 0.252141, 0.251809, 0.297989, 0.296257, 0.295319, 0.339603, 0.337731, 0.333820, 0.389510, 0.386998, 0.438864, 0.434887, 0.427893, 0.471568, 0.461699, 0.461144, 0.513532, 0.506641, 0.505062, 0.535648, 0.533726, 0.568064, 0.612886, 0.624169])
```

```python
f_mnq, temp = curve_fit(func, x_data, y_data)
f_mnq
```

```python
plt.plot(x_data,y_data, 'o')
plt.plot(x_data, func(x_data, *f_mnq))
plt.grid()
sum(func(x_data, *f_mnq))
```

### Построим эту же модель, но используя аппроксимацию полиномом

```python
p = np.polyfit(x_data, y_data, 2) #2 степень многочлена
f = np.poly1d(p)
fig = plt.figure()
plt.plot(x_data, y_data, 'o')
plt.plot(x_data,f(x_data))
plt.grid()
sum(f(x_data))
```
